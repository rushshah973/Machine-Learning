```python
import pandas as pd
```

```python
df=pd.read_csv('salaries.csv')
```

```python
df
```

```python
inputs=df.drop('salary_more_then_100k',axis='columns')
out=df.salary_more_then_100k
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
```

```python
inputs['le_company']=le_company.fit_transform(inputs['company'])
inputs['le_job']=le_company.fit_transform(inputs['job'])
inputs['le_degree']=le_company.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
ints=inputs.drop(['company','job','degree'],axis='columns')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(ints,out,test_size=0.2)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```
